## 1. Matplotlib ÁÆÄ‰ªã

Matplotlib ÊòØ Python ‰∏≠ÊúÄÂ∏∏Áî®ÁöÑ 2D ÁªòÂõæÂ∫ì„ÄÇÂÆÉÁöÑËÆæËÆ°Âì≤Â≠¶ÊòØÔºö**Áî®ÊúÄÂ∞ëÁöÑÂëΩ‰ª§ÁªòÂà∂Âá∫ÊúÄÂ∏∏ËßÅÁöÑÂõæÂΩ¢„ÄÇ**


```python

import numpy as np
import matplotlib.pyplot as plt
```


## 2. Âü∫Êú¨ÁªòÂõæÊµÅÁ®ãÔºàInitialize ‚Üí Prepare ‚Üí Render ‚Üí ObserveÔºâ


```python

# Initialize
import numpy as np
import matplotlib.pyplot as plt

# Prepare
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)

# Render
fig, ax = plt.subplots()
ax.plot(X, Y)

# Observe
plt.show()
```


## 3. Â∏∏ËßÅÂõæË°®Á±ªÂûãÔºàChooseÔºâ


```python

fig, axs = plt.subplots(3, 3, figsize=(10,10))

X = np.random.uniform(0, 1, 100)
Y = np.random.uniform(0, 1, 100)
axs[0,0].scatter(X, Y)
axs[0,0].set_title("Scatter")

X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
axs[0,1].bar(X, Y)
axs[0,1].set_title("Bar")

Z = np.random.uniform(0, 1, (8, 8))
axs[0,2].imshow(Z)
axs[0,2].set_title("Heatmap")

Z = np.random.uniform(0, 1, (8, 8))
axs[1,0].contourf(Z)
axs[1,0].set_title("Contour")

Z = np.random.uniform(0, 1, 4)
axs[1,1].pie(Z)
axs[1,1].set_title("Pie")

Z = np.random.normal(0, 1, 100)
axs[1,2].hist(Z)
axs[1,2].set_title("Histogram")

X = np.arange(5)
Y = np.random.uniform(0, 1, 5)
axs[2,0].errorbar(X, Y, Y/4)
axs[2,0].set_title("Errorbar")

Z = np.random.normal(0, 1, (100, 3))
axs[2,1].boxplot(Z)
axs[2,1].set_title("Boxplot")

axs[2,2].axis('off')

plt.tight_layout()
plt.show()
```


## 4. ÂõæÂΩ¢ÁæéÂåñÔºàTweakÔºâ


```python

X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, axs = plt.subplots(2, 2, figsize=(8,6))
axs[0,0].plot(X, Y, color='black')
axs[0,0].set_title("Color")

axs[0,1].plot(X, Y, linestyle='--')
axs[0,1].set_title("Linestyle")

axs[1,0].plot(X, Y, linewidth=5)
axs[1,0].set_title("Linewidth")

axs[1,1].plot(X, Y, marker='o')
axs[1,1].set_title("Marker")

plt.tight_layout()
plt.show()
```


## 5. ÂõæÂÉèÁªÑÁªáÔºàOrganizeÔºâ


```python

X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(X, Y1, color='C1')
ax2.plot(X, Y2, color='C0')
ax1.set_title("Sine")
ax2.set_title("Cosine")
plt.tight_layout()
plt.show()
```


## 6. Ê∑ªÂä†Ê†áÁ≠æÔºàLabelÔºâ


```python

X = np.linspace(0, 10, 100)
Y = np.sin(X)

fig, ax = plt.subplots()
ax.plot(X, Y)
fig.suptitle("Sine Wave Example")
ax.set_title("A Sine wave")
ax.set_xlabel("Time")
ax.set_ylabel("Amplitude")
plt.show()
```


## 7. ‰øùÂ≠òÂõæÂÉèÔºàSaveÔºâ


```python

fig, ax = plt.subplots()
ax.plot(np.linspace(0,10,100), np.sin(np.linspace(0,10,100)))
ax.set_title("Save Example")
fig.savefig("my-first-figure.png", dpi=300)
fig.savefig("my-first-figure.pdf")
plt.show()
```
